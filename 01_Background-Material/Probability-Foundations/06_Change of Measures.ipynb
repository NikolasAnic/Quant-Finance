# Change of Measure

## Overview

A **change of measure** involves using a positive random variable $Z$ to transform one probability measure $P'$ into another probability measure $P$. This is fundamental in financial modeling, where we change from the **real-world probability measure ($P'$)** to the **risk-neutral probability measure ($P$)** for pricing derivatives.

---

## Theorem 1.6.1: Constructing a New Measure

Let $(\Omega, \mathcal{F}, P')$ be a probability space, and let $Z$ be a non-negative random variable satisfying:

1. $Z > 0$ almost surely,  
2. $\mathbb{E}_{P'}[Z] = 1$.  

Define the new probability measure $P$ for any event $A \in \mathcal{F}$ as:

$$
P(A) = \int_A Z(w) \, dP'(w).
$$

This transforms the measure $P'$ into $P$ using $Z$, which adjusts the weights assigned to events.

---

### Definition: Radon-Nikodym Derivative

The random variable $Z$ is called the **Radon-Nikodym derivative** of $P$ with respect to $P'$ and is denoted as:

$$
Z = \frac{dP}{dP'}.
$$

#### Properties of $Z$:
1. $Z > 0$ almost surely under $P'$,
2. $\mathbb{E}_{P'}[Z] = 1$,
3. Expectations under $P$ are computed via:
   $$
   \mathbb{E}_P[X] = \mathbb{E}_{P'}[X \cdot Z].
   $$

---

## What Does This Mean?

The Radon-Nikodym derivative $Z$ acts as a **lens** or a **reweighting factor** that modifies probabilities:
- **$Z(w) > 1$**: The event $w$ becomes more likely under $P$ than under $P'$.
- **$Z(w) < 1$**: The event $w$ becomes less likely under $P$ than under $P'$.
- **$Z(w) = 1$**: The event $w$ has the same likelihood under both $P'$ and $P$.

---

## Theorem 1.6.2: Radon-Nikodym Theorem

### Statement

If $P'$ and $P$ are **equivalent probability measures** on $(\Omega, \mathcal{F})$ (i.e., they agree on which sets have probability zero), then there exists a random variable $Z > 0$ such that:

1. $\mathbb{E}_{P'}[Z] = 1$,  
2. For any event $A \in \mathcal{F}$:
   $$
   P(A) = \int_A Z(w) \, dP'(w).
   $$

The random variable $Z$ satisfies:

$$
Z = \frac{dP}{dP'}.
$$

### Intuition

- The Radon-Nikodym Theorem ensures that **any equivalent probability measure can be expressed using a Radon-Nikodym derivative**.
- The derivative $Z$ determines how probabilities under $P'$ are scaled to create $P$.

---

## Application in Financial Mathematics

In financial models, we:
1. Use $P'$ (real-world probabilities) to model observed market dynamics,
2. Use $P$ (risk-neutral probabilities) to price derivatives, ensuring no arbitrage.

### Key Insights:
- The change of measure adjusts probabilities but does not alter the underlying sample space or random variables.
- The relationship:
  $$
  \mathbb{E}_P[X] = \mathbb{E}_{P'}[X \cdot Z],
  $$
  allows us to compute expectations under $P$ using $P'$.

---

## Example: Risk-Neutral Pricing

Consider a random variable $X$ (e.g., the payoff of a derivative). Under $P'$ (real-world measure), the expectation reflects observed probabilities. To price $X$ under the risk-neutral measure $P$:
1. Define the Radon-Nikodym derivative $Z$ such that:
   $$
   \mathbb{E}_{P'}[Z] = 1 \quad \text{and} \quad Z > 0 \text{ almost surely}.
   $$

2. Compute probabilities and expectations under $P$:
   - Probability of an event $A$:
     $$
     P(A) = \int_A Z(w) \, dP'(w).
     $$
   - Expected value of $X$:
     $$
     \mathbb{E}_P[X] = \mathbb{E}_{P'}[X \cdot Z].
     $$

### Why Does This Work?
- $Z$ reweights probabilities, reflecting the shift from $P'$ to $P$.
- Prices derived under $P$ are arbitrage-free and consistent with the real-world dynamics described by $P'$.

---

## Conclusion

The Radon-Nikodym Theorem provides the mathematical foundation for changing probability measures. In financial contexts, it enables us to move from the real-world measure $P'$ to the risk-neutral measure $P$ while preserving the underlying probability structure. This ensures that derivative pricing remains consistent and arbitrage-free.
