# Chapter 1.7: Summary

## 1.7.1 Probability Space

Probability theory begins with the definition of a **probability space** $(\Omega, \mathcal{F}, P')$:
- $\Omega$: The set of all possible outcomes of a random experiment.
- $\mathcal{F}$: A $\sigma$-algebra of subsets of $\Omega$ where probabilities are defined.
- $P'$: A probability measure mapping $\mathcal{F}$ to $[0, 1]$.

The two key axioms of a probability space are:
1. $P'(\Omega) = 1$,
2. **Countable Additivity**: For any countable sequence of disjoint sets $\{A_i\}$ in $\mathcal{F}$:
   $$
   P'\left(\bigcup_{i} A_i\right) = \sum_i P'(A_i).
   $$

---

## 1.7.2 $\sigma$-Algebra and Borel Sets

- A **$\sigma$-algebra** $\mathcal{F}$ satisfies:
  - $\emptyset \in \mathcal{F}$,
  - If $A \in \mathcal{F}$, then $A^c \in \mathcal{F}$,
  - If $\{A_i\}_{i=1}^\infty \subset \mathcal{F}$, then $\bigcup_i A_i \in \mathcal{F}$.

- The **Borel $\sigma$-algebra** $\mathcal{B}(\mathbb{R})$ is the smallest $\sigma$-algebra containing all closed intervals $[a, b]$ in $\mathbb{R}$.

---

## 1.7.3 Random Variables and Distributions

- A **random variable** $X$ is a mapping $X: \Omega \to \mathbb{R}$ such that for every $B \in \mathcal{B}(\mathbb{R})$:
  $$
  \{\omega \in \Omega : X(\omega) \in B\} \in \mathcal{F}.
  $$

- The **distribution** of $X$ is a measure $\mu_X$ on $\mathbb{R}$, defined as:
  $$
  \mu_X(B) = P'(X \in B) \quad \text{for } B \in \mathcal{B}(\mathbb{R}).
  $$

### Special Cases:
1. **Density**: If $X$ has a density $f(x)$:
   $$
   \mu_X(B) = \int_B f(x) \, dx.
   $$

2. **Discrete Distribution**: If $X$ takes countably many values $\{x_i\}$:
   $$
   \mu_X(B) = \sum_{x_i \in B} P'(X = x_i).
   $$

---

## 1.7.4 Expectation and Integration

The **expectation** of a random variable $X$ is:
$$
\mathbb{E}[X] = \int_\Omega X(\omega) \, dP'(\omega),
$$
where the right-hand side is a **Lebesgue integral**.

### Properties of Lebesgue Integrals (Theorem 1.3.4):
1. **Comparison**: If $X \leq Y$ almost surely, then:
   $$
   \mathbb{E}[X] \leq \mathbb{E}[Y].
   $$

2. **Linearity**: For $a, b \in \mathbb{R}$:
   $$
   \mathbb{E}[aX + bY] = a\mathbb{E}[X] + b\mathbb{E}[Y].
   $$

3. **Jensen's Inequality**: For a convex function $\phi$:
   $$
   \phi(\mathbb{E}[X]) \leq \mathbb{E}[\phi(X)].
   $$

---

## 1.7.5 Monotone and Dominated Convergence

For a sequence of random variables $\{X_n\}$ converging almost surely to $X$:
1. **Monotone Convergence (Theorem 1.4.5)**:
   If $0 \leq X_1 \leq X_2 \leq \cdots \leq X_n$ almost surely, then:
   $$
   \lim_{n \to \infty} \mathbb{E}[X_n] = \mathbb{E}[X].
   $$

2. **Dominated Convergence (Theorem 1.4.9)**:
   If there exists a random variable $Y$ with $\mathbb{E}[|Y|] < \infty$ and $|X_n| \leq Y$ almost surely, then:
   $$
   \lim_{n \to \infty} \mathbb{E}[X_n] = \mathbb{E}[X].
   $$

---

## 1.7.6 Change of Measure

Suppose we have two probability measures $P'$ and $P$ on $(\Omega, \mathcal{F})$. A **change of measure** involves defining $P$ using a non-negative random variable $Z$ such that $\mathbb{E}_{P'}[Z] = 1$. For any $A \in \mathcal{F}$:
$$
P(A) = \int_A Z(\omega) \, dP'(\omega).
$$

### Key Results:
1. If $Z > 0$ almost surely, $P'$ and $P$ are **equivalent** (agree on sets with probability zero).
2. For any random variable $X$, the expectation under $P$ is:
   $$
   \mathbb{E}_P[X] = \mathbb{E}_{P'}[X \cdot Z].
   $$

---

## 1.7.7 Summary of Concepts

1. **Probability Space**: Defines the framework for random experiments.
2. **Random Variables**: Map outcomes to the real line, inducing distributions.
3. **Expectation**: Computed via Lebesgue integration, with useful properties like linearity and Jensen's inequality.
4. **Convergence Theorems**:
   - Monotone Convergence: Applies to non-decreasing sequences.
   - Dominated Convergence: Requires a dominating random variable.
5. **Change of Measure**: Used to transition between equivalent probability measures, essential in financial mathematics for modeling real-world and risk-neutral measures.
