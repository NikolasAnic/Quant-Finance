## Brownian motion

The **Brownian motion** $W(t)$ is the **limiting process** of a scaled symmetric random walk as the step size shrinks to zero while the number of steps increases indefinitely. This process is fundamental in probability theory and financial mathematics, modeling random continuous motion.

---

### **Definition of Brownian Motion**

A **stochastic process** $\{W(t)\}_{t \geq 0}$, defined on a probability space $(\Omega, \mathcal{F}, \mathbb{P})$, is called **Brownian motion** if it satisfies the following four properties:

#### **1. Initial Condition**

At time $t = 0$, the process starts at zero:
$$
W(0) = 0.
$$
This means that, regardless of the randomness, every sample path of the process begins at the origin.

#### **2. Independent Increments**

For any sequence of time points:
$$
0 = t_0 < t_1 < t_2 < \dots < t_m,
$$

the increments:

$$
W(t_1) - W(t_0), \quad W(t_2) - W(t_1), \quad \dots, \quad W(t_m) - W(t_{m-1})
$$
are **statistically independent**.

This property means that the value of the Brownian motion at future times does not depend on its past increments. That is, knowing how the process evolved up to some time $s$ provides no information about its future increments $W(t) - W(s)$, for $t > s$.

#### **3. Gaussian (Normal) Increments**

Each increment $W(t) - W(s)$, where $s < t$, is **normally distributed** with mean zero and variance proportional to the time difference:
$$
W(t) - W(s) \sim \mathcal{N}(0, t - s).
$$
This means that for any $0 \leq s < t$,
$$
\mathbb{E}[W(t) - W(s)] = 0, \quad \text{and} \quad \text{Var}[W(t) - W(s)] = t - s.
$$
Since the variance of an increment grows **linearly** with time, larger time intervals result in more uncertainty in the value of $W(t)$.

#### **4. Continuous Paths**
With probability one, the function $t \mapsto W(t)$ is **continuous**. This means that the graph of any sample path of Brownian motion has **no jumps**. However, it is also **nowhere differentiable**, meaning that the function has **infinite variation** over any time interval and cannot have a well-defined slope at any point.

This property is crucial in finance, as it suggests that asset prices modeled using Brownian motion cannot experience sudden discrete jumps (unless additional jump components are introduced, as in jump-diffusion models).

---

### **Brownian Motion as the Limit of the Scaled Random Walk**
Brownian motion arises as the **limit** of the **scaled symmetric random walk** as the number of steps increases indefinitely.

#### **Definition of the Scaled Random Walk**
Let's summarise the scaled random walk again which we defined in the previous chapter. Consider a symmetric random walk $M_n$, where each step is $\pm 1$ with equal probability. We define the **scaled random walk** as:

$$
W^{(n)}(t) = \frac{1}{\sqrt{n}} M_{nt}, \quad t \geq 0.
$$
where $M_{nt}$ counts the net number of heads minus tails in the first $nt$ coin flips.

By the **Central Limit Theorem**, for any fixed time $t$, the random variable $W^{(n)}(t)$ converges in distribution to a normal random variable:
$$
W^{(n)}(t) \xrightarrow{d} \mathcal{N}(0, t), \quad \text{as } n \to \infty.
$$
Thus, the **sequence of processes** $W^{(n)}(t)$ converges to **Brownian motion** $W(t)$ in the limit.

---

### **Key Differences Between Brownian Motion and the Scaled Random Walk**
| Property | Scaled Random Walk $W^{(n)}(t)$ | Brownian Motion $W(t)$ |
|----------|--------------------------------|----------------------|
| **Increments** | Approximately normal | Exactly normal |
| **Independence** | Increments are independent | Increments are independent |
| **Continuity** | Piecewise linear (jumps at discrete steps) | Continuous everywhere |
| **Differentiability** | Differentiable except at step points | Nowhere differentiable |
| **Variance Growth** | Proportional to time | Proportional to time |

Thus, Brownian motion provides a **continuous-time** model of random movement, which is particularly useful in **stochastic processes**, **financial modeling**, and **physics**.

---

### **Joint Normality and Alternative Characterizations of Brownian Motion**

#### **Joint Normality of Brownian Motion**
A key property of Brownian motion $W(t)$ is that for any set of times:
$$
0 \leq t_1 < t_2 < \dots < t_m,
$$
the random vector:
$$
(W(t_1), W(t_2), \dots, W(t_m))
$$
is **jointly normally distributed**.

This follows directly from the **independent normal increments** property of Brownian motion. Since each increment:
$$
W(t_i) - W(t_{i-1})
$$
is normally distributed, the entire vector $(W(t_1), W(t_2), \dots, W(t_m))$ follows a **multivariate normal distribution**.

#### **Mean and Covariance Structure**
Each $W(t_i)$ has an **expected value** of:
$$
\mathbb{E}[W(t_i)] = 0.
$$
For any two times $s < t$, the **covariance** of $W(s)$ and $W(t)$ is given by:
$$
\mathbb{E}[W(s) W(t)]
= \mathbb{E}[W(s) (W(t) - W(s))] + \mathbb{E}[W^2(s)]
= 0 + \text{Var}(W(s)) = s.
$$
Since $W(s)$ and $W(t) - W(s)$ are independent (by the independent increments property), their product expectation is zero.

Thus, the **covariance matrix** of $(W(t_1), W(t_2), \dots, W(t_m))$ is:
$$
\Sigma =
\begin{bmatrix}
t_1 & t_1 & t_1 & \dots & t_1 \\
t_1 & t_2 & t_2 & \dots & t_2 \\
t_1 & t_2 & t_3 & \dots & t_3 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
t_1 & t_2 & t_3 & \dots & t_m
\end{bmatrix}.
$$
This **covariance structure** shows that the correlation between $W(s)$ and $W(t)$ is given by:
$$
\text{Corr}(W(s), W(t)) = \frac{\text{Cov}(W(s), W(t))}{\sqrt{\text{Var}(W(s)) \text{Var}(W(t))}} = \sqrt{\frac{s}{t}}, \quad (s \leq t).
$$

#### **Moment-Generating Function Approach**
To confirm that $(W(t_1), W(t_2), ..., W(t_m))$ follows a **joint normal distribution**, we compute its **moment-generating function**:
$$
\varphi(u_1, u_2, ..., u_m) = \mathbb{E} \left[ \exp \left( \sum_{i=1}^{m} u_i W(t_i) \right) \right].
$$
By using the linearity of increments and their independence:
$$
\sum_{i=1}^{m} u_i W(t_i) = u_m (W(t_m) - W(t_{m-1})) + (u_{m-1} + u_m) (W(t_{m-1}) - W(t_{m-2})) + \dots + (u_1 + u_2 + ... + u_m) W(t_1).
$$
Since each increment is normally distributed, applying the moment-generating function of a normal variable leads to:
$$
\varphi(u_1, ..., u_m) = \exp \left( \frac{1}{2} \sum_{i=1}^{m} \left( \sum_{j=i}^{m} u_j \right)^2 (t_i - t_{i-1}) \right).
$$
This confirms that $(W(t_1), W(t_2), ..., W(t_m))$ follows a **joint normal distribution**.

---

### **Alternative Characterizations of Brownian Motion**
The theorem states that the following three properties are equivalent:

#### **(i) Independent Normal Increments**
For all time points:
$$
0 = t_0 < t_1 < \dots < t_m,
$$
the increments:
$$
W(t_1) - W(t_0), W(t_2) - W(t_1), \dots, W(t_m) - W(t_{m-1})
$$
are **independent** and normally distributed with:
$$
\mathbb{E}[W(t_i) - W(t_{i-1})] = 0, \quad \text{Var}[W(t_i) - W(t_{i-1})] = t_i - t_{i-1}.
$$
This definition is the most intuitive and is often taken as the **primary definition of Brownian motion**.

#### **(ii) Joint Normality**
The vector:
$$
(W(t_1), W(t_2), ..., W(t_m))
$$
is jointly normally distributed with **mean zero** and **covariance matrix**:
$$
\Sigma_{ij} = \mathbb{E}[W(t_i) W(t_j)] = \min(t_i, t_j).
$$
This formulation emphasizes that **Brownian motion is a Gaussian process**.

#### **(iii) Moment-Generating Function Characterization**
The **joint moment-generating function** of $(W(t_1), W(t_2), ..., W(t_m))$ is:
$$
\mathbb{E} \left[ \exp \left( \sum_{i=1}^{m} u_i W(t_i) \right) \right] = \exp \left( \frac{1}{2} \sum_{i=1}^{m} \sum_{j=1}^{m} u_i u_j \min(t_i, t_j) \right).
$$
This formulation is particularly useful in **proving joint normality**.

---

### **Filtration and Information in Brownian Motion**

In the study of **stochastic processes**, particularly **Brownian motion**, it is essential to define the **amount of information available at each time**. This is done using a **filtration**, which represents the growing set of information as time progresses.

A **filtration** is a **collection of $\sigma$-algebras** $\mathcal{F}(t)$, indexed by time $t \geq 0$, which satisfies the following three properties:

#### **(i) Information Accumulates Over Time**
For any two times $s$ and $t$ such that:
$$
0 \leq s < t,
$$
the $\sigma$-algebra at the earlier time must be contained within the $\sigma$-algebra at the later time:
$$
\mathcal{F}(s) \subseteq \mathcal{F}(t).
$$
This means that as time progresses, we **accumulate more information**—we do not lose any of the past information.

#### **(ii) Adaptivity**
For each time $t \geq 0$, the value of the **Brownian motion** at time $t$, denoted as $W(t)$, must be **$\mathcal{F}(t)$-measurable**. This means that:
$$
W(t) \text{ is completely determined by the information available at time } t.
$$
In other words, $\mathcal{F}(t)$ must include all information necessary to determine $W(t)$.

#### **(iii) Independence of Future Increments**
For any two times:
$$
0 \leq t < u,
$$
the **increment** of the Brownian motion:
$$
W(u) - W(t)
$$
is **independent** of the information contained in $\mathcal{F}(t)$. That is:
$$
W(u) - W(t) \perp \!\!\! \perp \mathcal{F}(t).
$$
This **ensures that future movements of Brownian motion cannot be predicted based on past information**. This is the **Markov property**, and it leads to the **efficient market hypothesis** in finance.


### **Adapted Stochastic Processes**
A **stochastic process** $\Delta(t)$ is said to be **adapted to the filtration** $\mathcal{F}(t)$ if, for each $t \geq 0$, the random variable $\Delta(t)$ is **$\mathcal{F}(t)$-measurable**. This means that at any given time $t$, the value of $\Delta(t)$ can be determined using the information in $\mathcal{F}(t)$.

Mathematically:
$$
\forall t \geq 0, \quad \Delta(t) \text{ is } \mathcal{F}(t)\text{-measurable}.
$$

### **Interpretation of Filtration in Asset Pricing**
In **financial modeling**, the filtration $\mathcal{F}(t)$ plays a crucial role:

- It represents **all publicly available information** at time $t$.
- The **efficient market hypothesis** states that all known information is already **incorporated into asset prices**.
- Since **future Brownian increments are independent of past information** (Property (iii)), this implies that **future price movements are unpredictable based on past prices alone**.

### **Two Possible Filtrations for Brownian Motion**
1. **Natural Filtration**: $\mathcal{F}(t)$ contains only the information obtained by observing **the Brownian motion itself** up to time $t$. This is the most common filtration and is written as:
   $$
   \mathcal{F}_t^W = \sigma(W(s) \mid 0 \leq s \leq t).
   $$
   This filtration **only includes past values of $W(t)$** and no external information.

2. **Enlarged Filtration**: $\mathcal{F}(t)$ contains information from observing **both the Brownian motion and other processes**. However, to **preserve the independence of future Brownian increments** (Property (iii)), this additional information **cannot reveal anything about future movements of $W(t)$**.

---

### **Explanation: Martingale Property for Brownian Motion**

A stochastic process \( W(t) \) is a **martingale** with respect to a filtration \( \mathcal{F}(t) \) if, for any two times \( 0 \leq s \leq t \),
$$
\mathbb{E}[W(t) \mid \mathcal{F}(s)] = W(s).
$$
This means that given all the information up to time \( s \), the expected future value of \( W(t) \) (for \( t > s \)) is simply its current value \( W(s) \). This confirms that Brownian motion has **no tendency to drift upwards or downwards** on average.


#### **Proof:**
We prove that Brownian motion satisfies the martingale property:

1. Start with the conditional expectation:
   $$
   \mathbb{E}[W(t) \mid \mathcal{F}(s)].
   $$

2. Using the definition of increments in Brownian motion:
   $$
   W(t) = (W(t) - W(s)) + W(s).
   $$

3. Taking conditional expectation:
   $$
   \mathbb{E}[W(t) \mid \mathcal{F}(s)]
   = \mathbb{E}[(W(t) - W(s)) + W(s) \mid \mathcal{F}(s)].
   $$

4. Using **linearity of expectation**:
   $$
   \mathbb{E}[W(t) \mid \mathcal{F}(s)] =
   \mathbb{E}[W(t) - W(s) \mid \mathcal{F}(s)] + \mathbb{E}[W(s) \mid \mathcal{F}(s)].
   $$

5. By the **independence of Brownian increments** (since \( W(t) - W(s) \) is independent of \( \mathcal{F}(s) \)):
   $$
   \mathbb{E}[W(t) - W(s) \mid \mathcal{F}(s)] = \mathbb{E}[W(t) - W(s)].
   $$

6. Since \( W(t) - W(s) \) is normally distributed with mean **zero**, we have:
   $$
   \mathbb{E}[W(t) - W(s)] = 0.
   $$

7. Therefore:
   $$
   \mathbb{E}[W(t) \mid \mathcal{F}(s)] = 0 + W(s) = W(s).
   $$

Thus, Brownian motion satisfies the **martingale property**.


#### **Interpretation**
- The **martingale property** means that **knowing the past does not help in predicting the future**.
- The expected future value of the Brownian motion is **equal to its current value**.
- This property is crucial in financial mathematics, especially in **option pricing and risk-neutral valuation**.

### **Quadratic Variation of Brownian Motion**

#### **Overview**

The **quadratic variation** of a function measures how much it fluctuates over time by summing the squared increments along a sequence of partitions. For **Brownian motion**, the quadratic variation plays a crucial role in **stochastic calculus**, distinguishing it from ordinary calculus. 

Unlike smooth functions, which have zero quadratic variation, the paths of Brownian motion accumulate a quadratic variation of **\(T\) over time \(T\)**. This property is essential in **Ito calculus** and contributes to the **volatility term** in the Black-Scholes-Merton model for option pricing.

### **First-Order Variation**
Before discussing quadratic variation, consider the **first-order variation**, which measures the total oscillation of a function.

1. Given a function f(t), its first-order variation up to time T is:
   $$
   FV_T(f) = \lim_{\| \Pi \| \to 0} \sum_{j=0}^{n-1} |f(t_{j+1}) - f(t_j)|.
   $$
   Here, $\Pi = \{t_0, t_1, \dots, t_n\}$ is a partition of [0,T], and $| \Pi \|$ is the maximum step size in the partition.

2. Using the **Mean Value Theorem**, we can rewrite it as:
   $$
   FV_T(f) = \int_0^T |f'(t)| dt.
   $$
   This formula holds for functions with well-defined derivatives.

3. If \( f(t) \) has a **continuous derivative**, then its **quadratic variation** is zero:
   $$
   [f,f](T) = \lim_{\| \Pi \| \to 0} \sum_{j=0}^{n-1} [f(t_{j+1}) - f(t_j)]^2 = 0.
   $$
   This is why quadratic variation is **not useful** for smooth functions.

### **Definition of Quadratic Variation**
For a function f(t), the **quadratic variation** up to time T is defined as:
   $$
   [f,f](T) = \lim_{\| \Pi \| \to 0} \sum_{j=0}^{n-1} [f(t_{j+1}) - f(t_j)]^2.
   $$
This measures the accumulated squared changes of f(t).

### **Key Idea: What Does Quadratic Variation Mean?**

The quadratic variation of a function measures how much its **square of the changes** (increments) add up over an interval. It tells us about the **roughness** or **irregularity** of a function's path.

For a smooth function $f(t)$, the sum of squared increments is small (often approaching **zero**), meaning it has **no quadratic variation**. However, for Brownian motion $W(t)$, the **sum of squared increments grows at the same rate as time**, meaning it has **positive quadratic variation**.

This means that if we observe Brownian motion from $t=0$ to $t=T$, the **sum of squared changes always adds up to $T$ in the limit**.

So for a **Brownian motion**, we will show that its quadratic variation is equal to **T**.


### **Key Result: Quadratic Variation of Brownian Motion**

#### **Theorem 3.4.3**: *Let W(t) be a Brownian motion. Then, almost surely,*
   $$
   [W, W](T) = T, \quad \text{for all } T \geq 0.
   $$
This means that the **quadratic variation of a Brownian motion grows linearly with time**.

### **Proof: Quadratic Variation of Brownian Motion**

We want to prove that for a **Brownian motion** $W(t)$, the **quadratic variation** up to time $T$ is:

$$
[W, W](T) = T, \quad \text{almost surely}.
$$

This means that if we sum the squared increments of the Brownian motion over finer and finer partitions of the interval $[0, T]$, the sum will converge to $T$.

### **Step 1: Definition of Quadratic Variation**
The **quadratic variation** of a function $W(t)$ on the interval $[0, T]$ is defined as:

$$
[W, W](T) = \lim_{\|\Pi\| \to 0} \sum_{i=0}^{n-1} \left[ W(t_{i+1}) - W(t_i) \right]^2,
$$

where $\Pi = \{0 = t_0 < t_1 < \dots < t_n = T\}$ is a **partition** of the interval $[0, T]$, and $\|\Pi\|$ is the **mesh size** (i.e., the length of the largest subinterval in the partition). The limit is taken as $\|\Pi\| \to 0$, meaning that the partition becomes finer and finer.

### **Step 2: Quadratic Variation for a Discrete Approximation**
To approximate $[W, W](T)$, we divide $[0, T]$ into **equally spaced** intervals of length $\frac{T}{n}$, meaning:

$$
t_i = \frac{iT}{n}, \quad \text{for } i = 0, 1, ..., n.
$$

The squared increments of $W(t)$ over this partition are:

$$
\sum_{i=0}^{n-1} \left[ W(t_{i+1}) - W(t_i) \right]^2.
$$

By the **independent increments property** of Brownian motion, we know:

$$
W(t_{i+1}) - W(t_i) \sim \mathcal{N}(0, \Delta t),
$$

where $\Delta t = \frac{T}{n}$ is the length of each subinterval.

Since Brownian motion has **independent increments**, the expected value of the sum of squared increments is:

$$
\mathbb{E} \left[ \sum_{i=0}^{n-1} \left(W(t_{i+1}) - W(t_i) \right)^2 \right] = \sum_{i=0}^{n-1} \mathbb{E} \left[ \left(W(t_{i+1}) - W(t_i) \right)^2 \right].
$$

Using the fact that $\text{Var}(W(t_{i+1}) - W(t_i)) = \Delta t$, we compute:

$$
\mathbb{E} \left[ (W(t_{i+1}) - W(t_i))^2 \right] = \frac{T}{n}.
$$

Summing over all $n$ subintervals:

$$
\mathbb{E} \left[ [W, W](T) \right] = \sum_{i=0}^{n-1} \frac{T}{n} = T.
$$

Thus, in expectation, $[W, W](T)$ accumulates at rate $T$, which suggests that in the limit, it should converge to $T$ **almost surely**.

### **Step 3: Convergence in Probability**
To establish almost sure convergence, we consider the **variance** of the sum:

$$
\text{Var} \left( \sum_{i=0}^{n-1} \left(W(t_{i+1}) - W(t_i) \right)^2 \right).
$$

Since the squared increments $ (W(t_{i+1}) - W(t_i))^2 $ are independent, their variance adds up:

$$
\text{Var} \left( (W(t_{i+1}) - W(t_i))^2 \right) = 2 \left(\frac{T}{n} \right)^2.
$$

Summing over all $n$ intervals:

$$
\text{Var} \left( [W, W](T) \right) = \sum_{i=0}^{n-1} 2 \left(\frac{T}{n} \right)^2 = 2 \frac{T^2}{n}.
$$

Taking the limit as $n \to \infty$, the variance goes to zero:

$$
\lim_{n \to \infty} \text{Var} \left( [W, W](T) \right) = 0.
$$

Since the variance goes to zero, the probability that $[W, W](T)$ deviates from $T$ by any positive amount also goes to zero, implying:

$$
[W, W](T) \to T, \quad \text{almost surely}.
$$

### **Intuition Behind the Proof**
Think of Brownian motion as a **random walk with infinitely small steps**. If we approximate it using a discrete-time random walk with small time steps, we see:

1. **Each small step contributes a squared increment** to the total sum.
2. **The steps are independent**—what happens at one step does not affect another.
3. **The variance of each step is proportional to its length**—longer steps contribute more to the sum.
4. **As the step size gets smaller, the total sum of squared increments approaches $T$**.

The intuition here is that the increments are not smooth, but are just continuous. Given that they are not smooth, their difference can diverge more than in a smooth function as the "individual steps" between two instances are more idiosyncratic and can thus be larger. As such, the quadratic difference will likely be larger than in a smooth function. Consequently, when we add more and more time steps to such a function, it is to be expected that it grows more rapidly. In terms of the **quadratic variation**, we were now able that, in its limit with the definitions above, the total sum of squared increments approaches T. 

We can visualize this through a comparison of a Brownian motion and a smooth function.


```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion(n, T):
    dt = T / n
    dW = np.random.normal(0, np.sqrt(dt), n)
    W = np.cumsum(dW)
    W = np.insert(W, 0, 0)  # Ensure W(0) = 0
    return W

def smooth_function(n, T):
    t = np.linspace(0, T, n+1)
    return np.sin(2 * np.pi * t)  # Example smooth function

# Parameters
n = 1000  # Number of time steps
T = 1.0   # Total time interval

t = np.linspace(0, T, n+1)
W = brownian_motion(n, T)
f = smooth_function(n, T)

# Compute quadratic variation
quad_var_W = np.cumsum(np.diff(W)**2)  # Brownian motion quadratic variation
quad_var_f = np.cumsum(np.diff(f)**2)  # Smooth function quadratic variation
quad_var_W = np.insert(quad_var_W, 0, 0)  # Ensure QV(0) = 0
quad_var_f = np.insert(quad_var_f, 0, 0)  # Ensure QV(0) = 0

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left panel: Function Comparison
axes[0].plot(t, W, label='Brownian Motion $W(t)$', color='blue')
axes[0].plot(t, f, label='Smooth Function $f(t) = \sin(2\pi t)$', linestyle='dashed', color='red')
axes[0].set_title("Comparison of $W(t)$ and $f(t)$")
axes[0].legend()

# Right panel: Quadratic Variation
axes[1].plot(t, quad_var_W, label='$[W, W](t)$ (Brownian Motion)', color='blue')
axes[1].plot(t, quad_var_f, label='$[f, f](t)$ (Smooth Function)', linestyle='dashed', color='red')
axes[1].set_title("Quadratic Variation")
axes[1].legend()

plt.tight_layout()
plt.show()
```

## Volatility of Geometric Brownian Motion

### Definition of Geometric Brownian Motion

A **Geometric Brownian Motion (GBM)** is defined as:

$$
S(t) = S(0) \exp \left( \sigma W(t) + \left( \mu - \frac{1}{2} \sigma^2 \right) t \right),
$$

where:
- $W(t)$ is a standard Brownian motion.
- $\sigma > 0$ is the **volatility** of the process.
- $\mu$ is the **drift** term.
- $S(0)$ is the initial stock price.

This model is the foundation of the **Black-Scholes-Merton option pricing formula**.

Here we show how to use the quadratic variation of Brownian motion to identify the volatility $\sigma$ from a path of this process.

#### Identifying Volatility Using Quadratic Variation

Given a time interval $[T_1, T_2]$, we **observe** the stock price $S(t)$ and compute the **log returns**:

$$
\log \left( \frac{S(t_{i+1})}{S(t_i)} \right) = \sigma \left(W(t_{i+1}) - W(t_i) \right) + \left( \mu - \frac{1}{2} \sigma^2 \right) (t_{i+1} - t_i).
$$

The **realized volatility** is computed by summing the squared log returns:

$$
\sum_{i=0}^{m-1} \left( \log \frac{S(t_{i+1})}{S(t_i)} \right)^2.
$$

Expanding this sum:

\begin{align}
\sum_{i=0}^{m-1} \left( \log \frac{S(t_{i+1})}{S(t_i)} \right)^2
&= \sigma^2 \sum_{i=0}^{m-1} (W(t_{i+1}) - W(t_i))^2 \\
&\quad + (\mu - \frac{1}{2} \sigma^2)^2 \sum_{i=0}^{m-1} (t_{i+1} - t_i)^2 \\
&\quad + 2\sigma (\mu - \frac{1}{2} \sigma^2) \sum_{i=0}^{m-1} (W(t_{i+1}) - W(t_i)) (t_{i+1} - t_i).
\end{align}

From properties of Brownian motion:

- **Quadratic variation** of $W(t)$ is $T_2 - T_1$, so:
  $$
  \sum_{i=0}^{m-1} (W(t_{i+1}) - W(t_i))^2 \approx (T_2 - T_1).
  $$
- **Quadratic variation** of $t$ is zero:
  $$
  \sum_{i=0}^{m-1} (t_{i+1} - t_i)^2 \to 0.
  $$
- **Cross variation** of $W(t)$ and $t$ is also zero:
  $$
  \sum_{i=0}^{m-1} (W(t_{i+1}) - W(t_i)) (t_{i+1} - t_i) \to 0.
  $$

Thus, when the partition step size is **small**, the squared log returns sum to:

$$
\sum_{i=0}^{m-1} \left( \log \frac{S(t_{i+1})}{S(t_i)} \right)^2 \approx \sigma^2 (T_2 - T_1).
$$

Taking the **square root**, we estimate the volatility as:

$$
\sigma \approx \sqrt{\frac{1}{T_2 - T_1} \sum_{i=0}^{m-1} \left( \log \frac{S(t_{i+1})}{S(t_i)} \right)^2 }.
$$


## Practical Considerations

- In an ideal world, we can **decrease the step size** infinitely to improve accuracy.
- In reality:
  - **Trades occur at discrete times**, limiting available data.
  - **Bid-ask spread** introduces noise in price movements at very small time intervals.
  - Market microstructure effects can distort short-term volatility estimates.

Despite these issues, **realized volatility** remains one of the most effective ways to estimate the **true volatility** of an asset in empirical finance.
